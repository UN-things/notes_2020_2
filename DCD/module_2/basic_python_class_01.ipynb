## ¿Qué es python?

* Python es un lenguage de programación orientado a objetos.
* Es interpretado, lo cuál implica que debe tenerse python instalado para poder ejecutar programas escritos en python.
* A diferencia de R no fue creado para el análisis de datos, aunque es uno de sus puntos mas fuertes en la actualidad.

## ¿Por qué es usado Python para analizar los datos?

* El salto de Python como lenguaje de datos se dió con la creación de múltiples librerías enfocadas al análisis de datos: Scipy, Numpy, Pandas, Matplotlib, [Plotly](https://plotly.com/), scikit-learn.
* Recientemente se ha convertido en el lenguaje por excelencia para temas de Deep Learning, con la llegada de librerías como Tensorflow, Keras, Theano, Pythorch. ([Top 23 Deep Learning](https://www.kdnuggets.com/2017/10/ranking-popular-deep-learning-libraries-data-science.html)).
* Es uno de los lenguajes de programación para los cuales Apache Spark tiene un API ([API python](https://spark.apache.org/docs/latest/api/python/index.html)).

* Grandes empresas en el mundo apelan a él para el análisis de datos: Google, Facebook, Instagram, Airbnb, Spotify y Netflix entre otros. Esto tiene como consecuencia el desarrollo de librerías para uso interno que luego son liberadas.
  * [Tensorflow](https://www.tensorflow.org/)(Google)
  * [Airflow](https://airflow.apache.org/) (Airbnb)
  * [Prophet](https://facebook.github.io/prophet/) (Facebook)



## Python vs R ¿Cuál es mejor?

* **Python y R no son enemigos**
* No hay uno que en general sea mejor que el otro
* Son excelentes complementos y cada uno tiene su enfoque puntual.
  - Python
      * Se entiende mejor con Deep Learning y Big Data.
      * Es de proposito general.
      * Tiene un mejor gestionamiento de memoria.
  - R
      * Se en tiende mejor con temas estríctamente estadísticos.
      * Para modelamiento de datos todo viene empaquetado dentro. de el "objeto" resultado.
