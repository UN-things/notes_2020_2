## Taller II

Cree una clase llamada rectangulo con las siguientes caracteristicas:

- El método __ init __ debe recibir alguno de los dos siguientes parámetros:
    - lista con vertices superior izquierdo en inferior derecho:
        - Ejemplo [(x_sup_izq, y_sup_izq),(x_inf_der, y_inf_der)]
    - Tupla con largo y ancho:
        - Ejemplo (largo, ancho) en este caso para constriur el cuadrado se toma como vertice inferior izquierdo la coordenada (0,0)

recibidos los parámetros el método init debe construir los 4 vértices como atributos 

- La clase rectangulo debe tener un método llamado calculo_area que almacene el área del rectángulo como un atributo

- La clase rectangulo debe tener un método llamado calculo_perimetro que almacene el perimetro del rectángulo como un atributo

- La clase rectangulo debe tener un método llamado escalar que reciba un flotante y escale el cuadrado dejando fijo el vertice izquierdo superior

- La clase rectangulo debe tener un método llamado desplazar que reciba diccionario de la siguiente forma y haga el respectivo desplazamiento:
    - Ejemplo {'x': 5, 'y': -3}, desplaza 5 unidades a la derecha y 3 unidades hacia abajo.
- La clase rectangulo debe tener el método __ str __ que imprima: vertices, área y périmetro con el formato de su preferencia

- Opcional: la clase debe tener una función escalar_desplazar que reciba un entero y un diccionario como en los items 4 y 5, y realice el escalamiento y el desplazamiento. Suregencia, esta función puede hacer un llamado a las ya existentes de los items 4 y 5.


```
class rectangulo():

  def __init__(self, param):
    # Dependiendo del valor que ingresa construímos nuestra clase vien sea por las dimensiones alto y ancho 
    # o por dos vertoces no adyacentes
    if isinstance(param, list):
      self.construccion_vertices(param)
    elif isinstance(param, tuple):
      self.construccion_dimensiones(param)
    self.calculo_area()
    self.calculo_perimetro()

  # Construcción de objeto por vertices no adyacentes
  def construccion_vertices(self, param):
    self.v_sup_izq = param[0]
    self.v_inf_der = param[1]
    self.v_sup_der = (self.v_inf_der[0], self.v_sup_izq[1])
    self.v_inf_izq = (self.v_sup_izq[0], self.v_inf_der[1])
    self.alto = self.v_sup_izq[1] - self.v_inf_der[1]
    self.ancho = self.v_inf_der[0] - self.v_sup_izq[0]

  # Construcción de objeto por alto y ancho
  def construccion_dimensiones(self, param):
    self.ancho, self.alto = param
    self.v_inf_izq = (0, 0)
    self.v_sup_izq = (0, self.alto)
    self.v_inf_der = (self.ancho, 0)
    self.v_sup_der = (self.ancho, self.alto)

  # Calculo del área
  def calculo_area(self):
    self.area = self.alto * self.ancho

  # Cálculo del perímetro
  def calculo_perimetro(self):
    self.perimetro = 2 * (self.alto + self.ancho) 

  # Escalamiento del rectangulo
  def escalar(self, escala):
    # Primero calcula los nuevos alto y ancho
    self.alto = self.alto * escala 
    self.ancho = self.ancho * escala

    # Con los nuevos alto y ancho re ajusta los vértices
    self.v_inf_izq = (self.v_inf_izq[0], self.v_sup_izq[1] - self.alto)
    self.v_inf_der = (self.v_inf_izq[0] + self.ancho, self.v_inf_izq[1])
    self.v_sup_der = (self.v_inf_der[0], self.v_sup_izq[1])
    self.calculo_area()
    self.calculo_perimetro()

  def desplazar(self, despl):
    # Nota: se repite mucho código, si se hubieran guardado los vertices en una lista o diccionario se iterarían 
    # haciéndolo más eficiente
    self.v_inf_izq = self.v_inf_izq[0] + despl['x'], self.v_inf_izq[1] + despl['y']
    self.v_sup_izq = self.v_sup_izq[0] + despl['x'], self.v_sup_izq[1] + despl['y']
    self.v_inf_der = self.v_inf_der[0] + despl['x'], self.v_inf_der[1] + despl['y']
    self.v_sup_der = self.v_sup_der[0] + despl['x'], self.v_sup_der[1] + despl['y']
    self.calculo_area()
    self.calculo_perimetro()

  # Se hace uso de dos métodos previamente definidos
  def escalar_desplaza(self, escala, despl):
    self.escalar(escala)
    self.desplazar(despl)
  
  # Se formatea el texto para imprimir el objeto
  def __str__(self):
    str_return = '''Los parámetros del actual rectangulo son:\n
    - Vertice superior izquierdo: {v_sup_izq}
    - Vertice superior derecho: {v_sup_der}
    - Vertice inferior izquierdo: {v_inf_izq}
    - Vertice inferior derecho: {v_inf_der}
    - Alto: {alto}
    - Ancho: {ancho}
    - Perímetro: {perimetro}
    - Área: {area}
    '''
    str_return = str_return.format(v_sup_izq=self.v_sup_izq,
                                   v_sup_der=self.v_sup_der,
                                   v_inf_izq=self.v_inf_izq,
                                   v_inf_der=self.v_inf_der,
                                   alto=self.alto,
                                   ancho=self.ancho,
                                   perimetro=self.perimetro,
                                   area=self.area)
    return str_return
```

```
# Se crea el recátngulo
tup = (5,8)
r1 = rectangulo(tup)
print(r1)

dc = {'x': 3, 'y': -1}
print('-'* 70)
print('Desplazamos el rectángulo con los parámetros {}'.format(dc))
print('-'* 70)
r1.desplazar(dc)
print(r1)

escala = 2
print('-'* 70)
print('Escalamos el rectángulo con en parámetro {}'.format(escala))
print('-'* 70)
r1.escalar(escala)
print(r1)

escala = 2
dc = {'x': 1, 'y': 1}
print('-'* 70)
print('Efectuamos de forma simultanea escalar y desplazar con los parámetros {} y {}'.format(escala, dc))
print('Nota: primero se escala y luego se desplaza')
print('-'* 70)
r1.escalar_desplaza(escala, dc)
print(r1)
```
